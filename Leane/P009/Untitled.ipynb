```python
import pandas as pd

df2003 = pd.read_csv("dadosMete-Brasilia/INMET_CO_DF_A001_BRASILIA_01-01-2003_A_31-12-2003.CSV", encoding='iso-8859-1', decimal=',', delimiter=';', skiprows=8)
df2003.head()
```

```python
df2013 = pd.read_csv("dadosMete-Brasilia/INMET_CO_DF_A001_BRASILIA_01-01-2013_A_31-12-2013.CSV", encoding='iso-8859-1', decimal=',', delimiter=';', skiprows=8)
df2013.head()
```

```python
df2023 = pd.read_csv("dadosMete-Brasilia/INMET_CO_DF_A001_BRASILIA_01-01-2023_A_31-12-2023.CSV", encoding='iso-8859-1', decimal=',', delimiter=';', skiprows=8)
df2023.head()
```

```python
from datetime import datetime

df2023["Hora UTC"] = df2023["Hora UTC"].apply(lambda x: datetime.strptime(x, "%H%M  %Z").strftime("%H:%M"))

df2023.head()
```

```python
df2023.rename(columns={'Data':'DATA (YYYY-MM-DD)', 'Hora UTC':'HORA (UTC)', 'RADIACAO GLOBAL (Kj/m²)': 'RADIACAO GLOBAL (KJ/m²)'}, inplace=True)
df = pd.concat([df2003, df2013, df2023])

df.head()
```

```python
data = df["DATA (YYYY-MM-DD)"].values

indice = pd.DatetimeIndex(data, name = "Data")
df.set_index(indice, inplace=True)
df
```

```python
df.drop("DATA (YYYY-MM-DD)", axis=1, inplace=True)
df
   
```

```python
modificacao = []
for columns in df.columns:
    if isinstance(columns, tuple):
        columns = columns[0]

    a = (columns,"")
    if(columns.find("PRESSAO")>-1 or columns.find("PRESSÃO")>-1):
        a = ("Pressão",columns)
    if(columns.find("TEMPERATURA")>-1):
        a = ("Temperatura",columns)
    if(columns.find("UMIDADE")>-1):
        a = ("Umidade",columns)
    if(columns.find("VENTO")>-1):
        a = ("Velocidade do Vento",columns)
    modificacao.append(a)
    
df.columns = pd.MultiIndex.from_tuples(modificacao)
df
```

Evolução das temperaturas médias e acumulado de precipitações nos tres
anos

```python
import matplotlib.pyplot as plt
import numpy as np

idf = df.groupby([df.index.year, df.index.month])["Temperatura"].mean()
idf = idf.loc[:]["Temperatura"]



fig, axes = plt.subplots(18, 1, figsize=(10, 40), sharex=True)
for ano, ax, columns in zip(['2003', '2013', '2023']*6, axes, np.concatenate((idf.columns,idf.columns,idf.columns))):
    idf.loc[ano,][columns].plot(ax=ax, title=ano, legend=True, xlabel="Dia", ylabel="Média")
plt.subplots_adjust(wspace=1, hspace=1)






```

```python
idf = df.groupby([df.index.year])["PRECIPITAÇÃO TOTAL, HORÁRIO (mm)"].cumsum()

fig, axes = plt.subplots(3, 1, figsize=(10, 10))

for ano, ax in zip(['2003', '2013', '2023'], axes):
    idf.loc[ano].plot(ax=ax, title=ano, ylabel="Precipitação Acumulada")
plt.subplots_adjust(wspace=1, hspace=1)
```

```python
import seaborn as sns

Ano = df.index.get_level_values("Data").year
idf = df.groupby([Ano, "Estação"])["Temperatura"].agg([ "min", "max"])
anos = [2003,2013,2023]

num_rows = len(idf.columns)
num_cols = 1

fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(10, 6*num_rows))

for ax, coluna in zip(axes, idf.columns):
    for ano in anos:
        idf.xs(ano, level="Data")[coluna].plot(ax=ax, style=".-", label=str(ano))

    ax.set_title(coluna[1:3])
    ax.legend(title='Ano')

plt.tight_layout()
plt.show()
```
