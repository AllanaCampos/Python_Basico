Exercício 2: Trabalhando com SeriesVamos criar um objeto Index utilizando a lista de identificador. Utilizando este objeto
vamos converter cada uma das listas, geradas no exercício anterior, em objetos da
classe Series . Com base nos objetos criados:

```python
index=pd.Index(identificador)
idades= pd.Series(idade, index=index)
idadeMedia= idades.mean()
idadeMax = idades.idxmax()
idadeMin = idades.idxmin()
print(f"Media de idades da equipe Cobrinhas: {idadeMedia}")
print(f"Membro mais jovem: {idadeMin} {idades[idadeMin]} anos")
print(f"Membro mais velho: {idadeMax} {idades[idadeMax]} anos")
```

```python
formacao = pd.Series(formação, index=index)
contagemFormados = formacao[formacao == 3].count()
contagemGranduandos = formacao[formacao.isin([1,2])].count()
contagemTecnicos = formacao[formacao.isin([0,1])].count()
print(f"Quantidade de Formados: {contagemFormados}")
print(f"Quantidade de Graduandos: {contagemGranduandos}")
print(f"Quantidade de Técnicos: {contagemTecnicos}")
```

```python
formacaoGeral = pd.Series(formaçãoGeral, index=index)
engenharia = formacaoGeral[formacaoGeral==0].count()
computacao = formacaoGeral[formacaoGeral==1].count()
print(f"Engenheiros: {engenharia}")
print(f"Computação: {computacao}")
```

Exercício 3: Trabalhando com DataFrames
Crie um DataFrame utilizando as informações coletadas. Utilize o mesmo Index do
exercício anterior e de nomes apropriados às colunas. Você pode criar o objeto a partir
das listas coletadas ou dos objetos Series . Veja como foi feito o tratamento de dados
faltan o.

```python
data = {
    'Identificador': identificador,
    'Idade': idade,
    'Formação': formação,
    'Formação Geral': formaçãoGeral,
    'Formação Específica': formaçãoEspecífica,
    'Andamento Graduação (%)': andamentoGraduação,
    'Tempo Formação (anos)': tempoFormação,
    'Experiência Prévia': experiênciaPrevia
}

df = pd.DataFrame(data)
df.set_index('Identificador', inplace=True)  #tornando a coluna 'Identificador' como índice.
df
```

```python
idadeMedia= df['Idade'].mean()
idadeMax = df['Idade'].idxmax()
idadeMin = df['Idade'].idxmin()
print(f"Media de idades da equipe Cobrinhas: {idadeMedia}")
print(f"Membro mais jovem: {idadeMin} {df.loc[idadeMin,'Idade']} anos")
print(f"Membro mais velho: {idadeMax} {df.loc[idadeMax,'Idade']} anos")
```

```python
contagemFormados = df[df['Formação'] == 3].shape[0]
contagemGraduandos = df[df['Formação'].isin([1, 2])].shape[0]
contagemTecnicos = df[df['Formação'].isin ([0,1])].shape[0]
print(f"Quantidade de Formados: {contagemFormados}")
print(f"Quantidade de Graduandos: {contagemGraduandos}")
print(f"Quantidade de Técnicos: {contagemTecnicos}")
```

```python
engenharia = df[df['Formação Geral'] == 0].shape[0]
computação = df[df['Formação Geral'] == 1].shape[0]
print(f"Engenheiros: {engenharia}")
print(f"Computação: {computação}")
```

Exercício 4: Trabalhando com MultiIndex
Agora vamos criar uma nova dimensão dos dados considerando cada trilha por
separado. Agrupe os índices por trilha Python, .Net e Java. Para testar o funcionamento
troque com colegas de outras equipes as listas coletadas e utilize elas para preencher os
dados das duas novas dimensões.

```python
Tic18Py = [["tic18Py07799", "Python", 24, 2, 0, "Engenharia Química", 76.94, None, True],
           ["tic18Py05996", "Python", 27, 1, 0, "Engenharia Química", 81.38, None, True],
           ["tic18Py06799", "Python", 24, 3, 1, "Ciência da Computação", None, 1, True],
           ["tic18Py08099", "Python", 24, 1, 0, "Engenharia Química", 75.58, None, True],
           ["tic18Py07100", "Python", 23, 2, 0, "Engenharia Elétrica", 90.16, None, True]
          ]

Tic18Java = [["tic18Java05599", "Java", 24, 2, 0, "Engenharia Química", 76.94, None, True],
           ["tic18Java04397", "Java", 26, 1, 1, "Ciência da Computação", 81.38, None, True],
           ["tic18Java04398", "Java", 25, 3, 1, "Ciência da Computação", None, 1, True],
           ["tic18Java04799", "Java", 24, 1, 0, "Engenharia Química", 75.58, None, True],
           ["tic18Java13201", "Java", 22, 2, 0, "Engenharia Elétrica", 90.16, None, True]
          ]

Tic18Net = [["tic18Net05598", "Net", 25, 3, 0, "Engenharia Química", None, 3, True],
           ["tic18Net04397", "Net", 26, 1, 1, "Ciência da Computação", 81.38, None, True],
           ["tic18Net04398", "Net", 25, 3, 1, "Ciência da Computação", None, 1, True],
           ["tic18Net04799", "Net", 24, 1, 0, "Engenharia Química", 75.58, None, True],
           ["tic18Net13201", "Net", 22, 2, 0, "Engenharia Elétrica", 90.16, None, True]
          ]

multiIndexDF = pd.DataFrame(Tic18Py + Tic18Java + Tic18Net,
                             
                             columns=['Identificador', 'Trilha', 'Idade', 'Formação', 'Formação Geral', 'Formação Específica', 'Andamento Graduação', 'Tempo Formação', 'Experiência Prévia'])

multiIndexDF
```

```python
idadeMediaPorTrilha = multiIndexDF.groupby('Trilha')['Idade'].mean()
idadeMaxTrilha = multiIndexDF.groupby('Trilha')['Idade'].max()
idadeMinTrilha = multiIndexDF.groupby('Trilha')['Idade'].min()
print("Media: ")
print(idadeMediaPorTrilha)
print("Minima: ")
print(idadeMinTrilha)
print("Maxima: ")
print(idadeMaxTrilha)
```

```python
contagemTrilha = multiIndexDF.groupby(['Trilha', 'Formação'])['Identificador'].count().reset_index()
contagemFormados = contagemTrilha[contagemTrilha['Formação'] == 3].set_index('Trilha')['Identificador']
contagemGraduandos = contagemTrilha[contagemTrilha['Formação'].isin([1, 2])].groupby('Trilha')['Identificador'].sum()
contagemTecnicos = contagemTrilha[contagemTrilha['Formação'].isin([0, 1])].groupby('Trilha')['Identificador'].sum()

print("Contagem de Formados por Trilha:")
print(contagemFormados)
print("\nContagem de Graduandos por Trilha:")
print(contagemGraduandos)
print("\nContagem de Técnicos por Trilha:")
print(contagemTecnicos)
```

```python
engenheiros = multiIndexDF[multiIndexDF['Formação Geral'] == 0].groupby('Trilha').size()
computacao = multiIndexDF[multiIndexDF['Formação Geral'] == 1].groupby('Trilha').size()
print(f"Engenheiros: {engenheiros}")
print(f"Computação: {computacao}")
```
