O modulo random do numpy oferece a função randint que gera um ndarray de numeros inteiros dentro do intervalo semiaberto [valorMinimo, valorMaximo)
e para gerar numeros de ponto flutuante no intervalo fechado [valorMinimo, valorMaximo] utilizamos a função uniform

```python
import numpy as np

def geraIdadesNP(n, iMin = 18, iMax = 64):
    return np.random.randint(iMin,iMax+1,n)

print(geraIdadesNP(10))

def geraSalariosNP(n, sMin = 1320, sMax = 13200):
    return np.random.uniform(sMin,sMax,n)

print(geraSalariosNP(10))
```
