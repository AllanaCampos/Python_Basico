Exercício 2: Juntando os datasets

Padronizando as colunas

```python
brasilia2023.rename(columns={'Data':'DATA (YYYY-MM-DD)', 'Hora UTC':'HORA (UTC)', 'RADIACAO GLOBAL (Kj/m²)': 'RADIACAO GLOBAL (KJ/m²)'}, inplace=True)
brasilia2023['DATA (YYYY-MM-DD)'] = pd.to_datetime(brasilia2023['DATA (YYYY-MM-DD)'], format="%Y/%m/%d", errors='coerce')

```

Gerando MultiIndex de Ano, Meses e Dias

```python
brasilia = pd.concat([brasilia2003, brasilia2013, brasilia2023])
brasilia['DATA (YYYY-MM-DD)'] = pd.to_datetime(brasilia['DATA (YYYY-MM-DD)'])
colunaData = pd.DatetimeIndex(brasilia['DATA (YYYY-MM-DD)'])
brasilia['Ano'] = colunaData.year
brasilia['Mes'] = colunaData.month
brasilia['Dia'] = colunaData.day
#print(brasilia.dtypes)
brasilia.set_index(['Ano', 'Mes', 'Dia'], inplace=True)
brasilia.drop(['DATA (YYYY-MM-DD)'], axis=1, inplace=True)

brasilia
```

Multindex com precipitações,
pressão, radiação, temperatura, umidade e velocidade do vento

```python
colunas = brasilia.columns

palavrasChave = {
    'Precipitação': ['precipitação'],
    'Pressão': ['pressão', 'pressao'],
    'Radiação': ['radiação', 'radiacao'],
    'Temperatura': ['temperatura'],
    'Umidade': ['umidade'],
    'Vento': ['vento']
}
colunasSelecionadas = {categoria: [] for categoria in palavrasChave}

for coluna in colunas:
    for categoria, palavras in palavrasChave.items():
        if any(palavra in str(coluna).lower() for palavra in palavras):
            colunasSelecionadas[categoria].append(coluna)

brasiliaMindex = brasilia[colunasSelecionadas['Precipitação'] + colunasSelecionadas['Pressão'] +
                             colunasSelecionadas['Radiação'] + colunasSelecionadas['Temperatura'] +
                             colunasSelecionadas['Umidade'] + colunasSelecionadas['Vento']]

categorias = [categoria for categoria, colunaCategoria in colunasSelecionadas.items() for _ in colunaCategoria]
subcategorias = [coluna for colunaCategoria in colunasSelecionadas.values() for coluna in colunaCategoria]

multiindex = pd.MultiIndex.from_arrays([categorias, subcategorias], names=['', ''])

brasiliaMindex.columns = multiindex
brasiliaMindex
```
