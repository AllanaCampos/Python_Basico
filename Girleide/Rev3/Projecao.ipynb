Exercício 1: Projeção de salarios / gastos com abono

```python
class Pagamentos:

    def __init__(self):
        self.__salario = None
        self.__salarioAdicionado = None
        self.__adicionaSalario = None
        self.__listaSalarios = []
        self.__gastosAbono = []

    def entradaDados(self):
        self.__salario = int(input('Digite o salário a ser acrescentado:'))
        if self.__salario == 0:
            return True
        return False

    def adicionaSalario(self):
        if self.__salario * 0.20 <= 100:
            self.__adicionaSalario = 100
            self.__gastosAbono.append(self.__adicionaSalario)
            self.__salarioAdicionado = self.__salario + self.__adicionaSalario

        else:
            self.__adicionaSalario = (self.__salario * 0.20)
            self.__gastosAbono.append(self.__adicionaSalario)
            self.__salarioAdicionado = self.__salario + self.__adicionaSalario

        self.__listaSalarios.append(self.__salarioAdicionado)

    def mostraSalarios(self):
        print('lista de salarios', self.__listaSalarios)

    def numeroFuncionarios(self):
        print('numero de funcionários cadastrados', len(self.__listaSalarios))

    def gastoTotal(self):
        print('gasto total', sum(self.__gastosAbono))

    def salariosPiso(self):
        I=0
        for salario in self.__gastosAbono:
            if salario == 100:
                I+=1
        print('salarios no piso', I)
    
    def main(self):

        while True:
            opcao = int(input("Adicionar um novo salario? 1-sim, 2-não"))
            if opcao == 1:
                self.entradaDados()
                self.adicionaSalario() 
                self.salariosPiso()
                self.mostraSalarios()
                self.numeroFuncionarios()
                self.gastoTotal()
            else:
                break
```

```python
  
p = Pagamentos()
p.main()
```
