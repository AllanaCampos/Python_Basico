```python
from faker import Faker
import pandas as pd
import random
import numpy as np 
import matplotlib.pyplot as plt

faker = Faker('pt_BR')
```

Criando Data Frame

```python
nomeS = []
cpfS =[]
idadeS = []
sexoS = []
emailS =[]
notaEnem = []
cra_2 = np.full(1000, np.nan)
cra_4 = np.full(1000, np.nan)
cra_6 = np.full(1000, np.nan)
abandonoS = []
semestreAbandono = []

```

```python
for i in range(1000):

    nome = faker.name()
    nomeS.append(nome)
    
    cpf = faker.cpf()
    cpfS.append(cpf)
    
    idade = np.random.randint(18, 28)
    idadeS.append(idade)
    
    sexo = random.choice(['M','F'])
    if sexo == 'M':
        faker.name_male()
    else:
        faker.name_female
    sexoS.append(sexo)
    
    email = faker.email()
    emailS.append(email)
    
    enem = np.random.randint(640, 800)
    notaEnem.append(enem)
    
    abandono = faker.boolean()
    abandonoS.append(abandono)
    if abandono == True:
        semestre = np.random.randint (1,8)
        if semestre > 2:
            cra_2 [i] = np.random.uniform (5,10)
        elif semestre > 4:
            cra_4 [i] = np.random.uniform (5,10)
        elif semestre > 6:
            cra_6 [i] = np.random.uniform (5,10)
    else:
        semestre = 8
        cra_2 [i] = np.random.uniform (5,10)
        cra_4 [i] = np.random.uniform (5,10)
        cra_6 [i] = np.random.uniform (5,10)
    semestreAbandono.append(semestre)
    
```

```python
dados = ({
    
        "NOME": nomeS,
        "CPF": cpfS,
        "IDADE" : idadeS,
        "SEXO" : sexoS,
        "E-MAIL" : emailS,
        "NOTA ENEM": notaEnem,
        "CRA 2°": cra_2,
        "CRA 4°": cra_4,
        "CRA 6°": cra_6,
        "ABANDONO": abandonoS,
        "SEMESTRE DE ABANDONO": semestreAbandono,
    })

df = pd.DataFrame(dados)
df
```

Gerando CSV

```python
df.to_csv("DFaula1901.csv")
```

1- Identifique qual q relação de alunos de sexo masculino e feminino no conjunto.

```python
df = pd.read_csv("DFaula1901.csv")
df.head()
```

```python
relacaoMasFem = df["SEXO"].value_counts()
print(relacaoMasFem)
relacaoMasFem.plot(kind='barh', stacked=True, color="orange")
```

2- Dado que os alunos tem idade entre 18 e 28 anos, mostre qual a média da nota do ENEM dos alunos agrupados por idade.

```python
notaMedia = df.groupby("IDADE")["NOTA ENEM"].mean()
print(notaMedia)
notaMedia.plot(kind='barh', stacked=True, color="orange", xlabel="MÉDIA DO ENEM")
```

3 - Normalize as notas do ENEM de todos os alunos para um valor entre 5 e 10.

```python
df["NOTA ENEM"] = 5*(df["NOTA ENEM"]-640)/(800-640)+5
df.rename(columns={"NOTA ENEM":"NOTA ENEM NORMALIZADA"},inplace=True)
df
```
